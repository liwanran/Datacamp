## 2) The np.random module and Bernoulli trials: e.g. a flip of a possibly biased coin.
Let's say a bank made 100 mortgage loans. It is possible that anywhere between 0 and 100 of the loans will be defaulted upon. You would like to know the probability of getting a given number of defaults, given that the probability of a default is p = 0.05. 
<p>(1) You will perform 100 Bernoulli trials using the f(n, p). 
<br>(2) You will do this for another 100 Bernoulli trials. And again and again until we have tried it 1000 times. <br>(3) Then, you will plot a histogram describing the probability of the number of defaults. Include the normed=True keyword argument so that the height of the bars of the histogram indicate the probability.
<p> f( ) accepts two arguments: the number of trials n - in this case 100, the probability of success p - in this case the probability of a default, which is 0.05. On each iteration of the loop store the result in an entry of h.

```python
"""Perform n Bernoulli trials with success probability p and return number of successes."""
def f(n, p):
    e=0
    for i in range(n):
        g = np.random.random()
        if g < p:
            e += 1
    return e

np.random.seed(42)
h = np.empty(1000)
for i in range(1000):
    h[i] = f(100, 0.05)

plt.hist(h, normed=True)
plt.xlabel('Number of defaults out of 100 loans')
plt.ylabel('Probability')
plt.show()
```

### Compute the number of 100-loan simulations with 10 or more defaults: n_lose_money

```python
defaults = np.sum(h>=10)
defaults
```

### Compute and print probability of losing money

```python
defaults/len(h)
```

### Will the bank fail?

```python
def f(m):
    n = len(m)
    x=np.sort(m)
    y=np.arange(1, n+1)/n
    return x, y

x, y = f(h)
plt.plot(x, y, marker='.', linestyle='none')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

# 3. Probability distributions and stories: The Binomial distribution
### PMF: probability mass function. The set of probabilities of discrete outcomes

```python
# two arguments: the # of Bernoulli trials (coin flips) and the probability of success (heads).
np.random.binomial(4, 0.5)
```

```python
# size=10: specifies how many random numbers to sample out of the binomial distribution.
np.random.binomial(4, 0.5, size=10)
```

### We do 60 Bernoulli trials with a probability of success of 0.1. If we do this over and over and tally the results, we can plot the PMF.

```python
samples = np.random.binomial(60, 0.1, size=10000)
plt.hist(samples, normed=True)
plt.show()
```

```python
def f(m):
    n = len(m)
    x=np.sort(m)
    y=np.arange(1, n+1)/n
    return x, y

sns.set()
x, y = f(samples)
plt.plot(x, y, marker='.', linestyle='none')
plt.margins(0.02)
plt.xlabel('Number of successes')
plt.ylabel('CDF')
plt.show()
```

# 4. Poisson processes and the Poisson distribution
## Poisson process: the timing of the next event is completely independent of when the previous event happened. The Poisson distribution has 1 parameter: the average number of arrivals in a given length of time.
## Poisson distribution: a limit of the binomial distribution for rare events.
## np.random.poisson( ): include the size key argument to allow multiple samples.

```python
samples = np.random.poisson(6, size=10000)
x, y = f(samples)
plt.plot(x, y, marker='.', linestyle='none')
plt.margins(0.02)
plt.xlabel('Number of successes')
plt.ylabel('CDF')
plt.show()
```
