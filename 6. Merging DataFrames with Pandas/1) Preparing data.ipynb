## 1) Merging datasets: using a loop

```python
doc1 = ['datasets/sales-jan-2015.csv', 'datasets/sales-feb-2015.csv', 'datasets/sales-mar-2015.csv']
df1 = []
for i in doc1:
    df1.append(pd.read_csv(i))
type(df1)
```

```python
df1
```

### 2) Merging datasets: using a comprehension

```python
doc2 = ['datasets/sales-jan-2015.csv', 'datasets/sales-feb-2015.csv', 'datasets/sales-mar-2015.csv']
df2 = [pd.read_csv(i) for i in doc2]
type(df2)
```

```python
df2
```

## 3) Merging datasets: using glob. * is like a wild card.

```python
doc3 = glob('datasets/sales*.csv')
df3 = [pd.read_csv(i) for i in doc3]
type(df3)
```

```python
df3
```

## 2) Absolute temperature range

```python
week_range = w.loc['2013-7-1':'2013-7-7', ['Min TemperatureF','Max TemperatureF']]
week_range
```

```python
week_avg = w.loc['2013-7-1':'2013-7-7', 'Mean TemperatureF']
week_avg
```

```python
week_range/week_avg
```
